```python
import pandas  as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data_Set1.csv")
df.head()
```

```python
inputs = df.drop('slaves', axis= 'columns')
target = df['slaves']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
target

```

```python
le_distance = LabelEncoder()
le_people = LabelEncoder()
le_battery = LabelEncoder()
le_network = LabelEncoder()
le_time = LabelEncoder()
le_slave = LabelEncoder()
```

```python
inputs['distance_n'] = le_distance.fit_transform(inputs['distance'])
inputs['people_n'] = le_distance.fit_transform(inputs['people'])
inputs['battery_n'] = le_distance.fit_transform(inputs['battery'])
inputs['network_n'] = le_distance.fit_transform(inputs['network'])
inputs['time_n'] = le_distance.fit_transform(inputs['time'])

inputs.head()
```

```python
inputs_n = inputs.drop(['distance','people','battery','network','time'], axis = 'columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model= tree.DecisionTreeClassifier(criterion='entropy')
```

```python
fig, ax = plt.subplots(figsize=(50, 24))
tree.plot_tree(model.fit(inputs_n,target), fontsize=8)
plt.savefig('tree_high_dpi', dpi=150)
```
